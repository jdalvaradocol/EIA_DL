```python
import pandas as pd
import numpy as np

from keras.models import Sequential
from keras.layers import Dense
```

```python
dataset = pd.read_csv('/content/sample_data/diabetes_train.csv')
dataset
```

```python
X = dataset.iloc[:,:8].values
print(X)

Y = dataset.iloc[:,8].values
print(Y)
```

```python
model = Sequential()

## capa de entrada
model.add(Dense(15,activation="relu",input_dim = 8))

## capa oculta

model.add(Dense(10,activation="relu"))
model.add(Dense(5,activation="relu"))

## capa salida
model.add(Dense( 1 ,activation="sigmoid"))

model.compile(loss="binary_crossentropy",
              optimizer= 'adam',
              metrics= ['accuracy']
              )

red = model.fit(X , Y, batch_size = 10, epochs = 500)
```

```python

import matplotlib.pyplot as plt

plt.xlabel("Epocas")
plt.ylabel("Perdidas")
plt.plot(red.history["loss"])
```

```python
dataset_test = pd.read_csv('/content/sample_data/diabetes_test.csv')
dataset_test

X_data = dataset_test.iloc[:,:8].values
print(X_data)
Y_data = dataset_test.iloc[:, 8].values
print(Y_data)
```

```python
model.evaluate(X_data, Y_data)
```

```python
y_eval = model.predict(X_data)
y_red = np.round(y_eval)

Resultados = pd.DataFrame(y_red, columns=['Salida RNA'])
Resultados['Salida real'] = pd.DataFrame(Y_data)

Resultados
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import seaborn as sns

matriz = confusion_matrix(Resultados['Salida real'], Resultados['Salida RNA'])

disp = ConfusionMatrixDisplay(matriz)

disp.plot()
```
